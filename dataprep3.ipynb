```python
import pandas as pd
df=pd.read_csv("datasets/Travel details dataset1.csv")
```

```python
df['Destination']=df['Destination'].str.split(',').str[0]
df
```

```python
df = pd.get_dummies(df, columns=['Destination'], prefix='Destination')
df
```

```python
df['Accommodation cost'] = df['Accommodation cost'].str.replace('$', '')
```

```python
df['Transportation cost'] = df['Transportation cost'].str.replace('$', '')
```

```python
df['Transportation cost'] = df['Transportation cost'].str.replace('USD', '')
```

```python
df['Accommodation cost'] = df['Accommodation cost'].str.replace('USD', '')
```

```python
df.dtypes
```

```python
#df['Accommodation cost'] = pd.to_numeric(df['Accommodation cost'])
df['Accommodation cost'] = df['Accommodation cost'].str.replace(',', '')
```

```python
df['Accommodation cost'] = pd.to_numeric(df['Accommodation cost'])
```

```python
df['Transportation cost'] = df['Transportation cost'].str.replace(',', '')
```

```python
df['Transportation cost'] = pd.to_numeric(df['Transportation cost'])
df.dtypes
```

```python
df.dropna(inplace=True)
```

```python
#df = df.applymap(lambda x: 1 if x else 0)
#df = df.applymap(lambda x: 1 if isinstance(x, bool) else x)
bool_columns = df.select_dtypes(include=bool).columns
df[bool_columns] = df[bool_columns].map(lambda x: 1 if x else 0)
```

```python
df['Accommodation type'].unique()
```

```python
df = pd.get_dummies(df, columns=['Accommodation type'], prefix='Accommodation type', dtype=int)
df['Transportation type'].unique()
```

```python
df=pd.get_dummies(df, columns=['Transportation type'], prefix='Transportation', dtype=int)
```

```python
#df.rename(columns={'Transportation cost':'Transportation Typecost_'}, inplace=True)
#df.rename(columns={'Transportation_Car rental': 'Transportation_Car Rental'}, inplace=True)
df_cleaned = df.select_dtypes(exclude=['object'])
df = df_cleaned
drop_col=['Trip ID', 'Traveler age', 'Accommodation cost', 'Transportation cost']
df.drop(columns= drop_col, inplace=True)
x=df.drop(['Total cost'], axis=1) 
```

```python
y=df['Total cost']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
svm=SVR()
svm.fit(x_train,y_train)
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
gr=GradientBoostingRegressor()
gr.fit(x_train,y_train)

y_pred1 = lr.predict(x_test)
y_pred2 = svm.predict(x_test)
y_pred3 = rf.predict(x_test)
y_pred4 = gr.predict(x_test)
df1=pd.DataFrame({'Actual':y_test, 'LinearRegression':y_pred1, 'SVR':y_pred2, 'RandomForestRegressor':y_pred3, 'GradientBoostingRegressor':y_pred4})
```

```python
df1
```

```python
import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.plot(df1['Actual'], label='Actual')
plt.plot(df1['LinearRegression'], label='LinearRegression')
plt.legend()

plt.subplot(2,2,2)
plt.plot(df1['Actual'], label='Actual')
plt.plot(df1['SVR'], label='SVR')
plt.legend()

plt.subplot(2,2,3)
plt.plot(df1['Actual'], label='Actual')
plt.plot(df1['RandomForestRegressor'], label='RandomForestRegressor')
plt.legend()

plt.subplot(2,2,4)
plt.plot(df1['Actual'], label='Actual')
plt.plot(df1['GradientBoostingRegressor'], label='GradientBoostingRegressor')
plt.legend()
plt.tight_layout()
```

```python
from sklearn import metrics
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
print(score1,score2,score3,score4)
```

```python
s1 = metrics.mean_absolute_error(y_test,y_pred1)
s2 = metrics.mean_absolute_error(y_test,y_pred2)
s3 = metrics.mean_absolute_error(y_test,y_pred3)
s4 = metrics.mean_absolute_error(y_test,y_pred4)
print(s1,s2,s3,s4)
```
